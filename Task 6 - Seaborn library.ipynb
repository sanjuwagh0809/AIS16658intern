#Seaborn
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics. Seaborn is built on top of matplotlib and integrates closely with pandas data structures.

```python
# To utilize Seaborn,you need to have it installed in your Python environment.You can install it via pip:
# pip install seaborn 
# Once installed, you can import it in your Python script or Jupyter Notebook:
import seaborn as sns
#Then , you can start using its functions to create beautiful visualization of your data.
```

```python
# Scatter Plot : visualizing the relationship between two variables.
import seaborn as sns
import matplotlib.pyplot as plt
# load example dataset
tips = sns.load_dataset("tips")
#scatter plot
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
#line plot : Visualizing trends over time or other continuous variables.
#line plot
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.lineplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
# Bar plot : Visualizing the distribution of a categorical variable.
#Bar plot
sns.barplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
# Box plot : Visualizing the distribution of a continuous variable , often across different categories.
#Box plot
sns.boxplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
# Violin plot: Similar to a Similar to a box plot, but provides a more detailed view of the distribution.
#violin plot
sns.violinplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
# Pair plot : Visualizing pairwise relationships between variables in a dataset.
#pair plot
sns.pairplot(tips)
plt.show()
```

```python
#Heatmap:Visualizing the correlation matrix of variables.
#Heatmap
import pandas as pd
tips_numeric = pd.get_dummies(tips,drop_first=True)
c = tips_numeric.corr()
sns.heatmap(c,annot=True,cmap="rainbow")
plt.show()
```
